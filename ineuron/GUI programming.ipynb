```python
from tkinter import *
root = Tk()

def show():
    a = e1.get()
    b = e2.get()
    c = a +" "+b
    l3['text'] = c
    
def finish():
    root.destroy()


l1 = Label(root,text = "First Name")
l2 = Label(root,text = "Last Name")
e1 = Entry(root)
e2 = Entry(root)
l3 = Label(root,font = "Arial 10 bold")
b1 = Button(root,text = "Show",command = show)
b2 = Button(root,text = "Quit",command = finish)

l1.grid(row = 0 ,column = 0)
l2.grid(row = 1 ,column = 0)
e1.grid(row = 0 ,column = 1)
e2.grid(row = 1 ,column = 1)
l3.grid(row = 2 ,column = 0)
b1.grid(row = 3,column = 0)
b2.grid(row = 3,column = 1)


root.mainloop()
```

```python
from tkinter import *
root = Tk()

def divide():
    a = int(e1.get())
    b = int(e2.get())
    c = a / b
    l3['text'] = c
    
def quit():
    root.destroy()
    


l1 = Label(root,text = "Enter the first No.")
l2 = Label(root,text = "Enter the second No.")
l3 = Label(root,font = "Arial 11 bold")
e1 = Entry(root)
e2 = Entry(root)
b1 = Button(root,text = "Divide",command = divide)
b2 = Button(root,text = "Exit",command = quit)


l1.grid(row = 0 ,column = 0)
l2.grid(row = 1 ,column = 0)
e1.grid(row = 0 ,column = 1)
e2.grid(row = 1 ,column = 1)
l3.grid(row = 2 ,column = 0)
b1.grid(row = 3,column = 0)
b2.grid(row = 3,column = 1)


root.mainloop()
```
