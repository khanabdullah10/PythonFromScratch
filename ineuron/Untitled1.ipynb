```python
df = pd.read_csv('titanic.csv')
```

```python
df.dropna(axis = 1)
# wherever the NaN value  is
# it will be delete with the whole column as axis = 1(for row) and axis = 0 (for column and by default it is 0)
```

```python
df
```

```python
#Sum of Nan value inside a column
df.isnull().sum()
```

```python
#Sum of Nan value inside a column
df.isna().sum()
```

```python
# Drop NaN values if the thresholod is 800 ie. the values inside the column should be non NaN values
df.dropna(axis= 1, thresh = 800)
```

```python
# Drop NaN values if the how is all ie. whole column should be NaN values
df.dropna(axis = 1,how = "all")
```

```python
# # Drop NaN values if the how is any ie. wherever if the column has NaN values it will be deleted 
df.dropna(axis = 1, how = 'any')
```

```python
# filling the values of NaN with specific name
df.fillna(value = 'none')
```

```python
df
```

```python
# Wherever the NaN value is ,it will be replace with the mean(average)  of Age column
df.fillna(value = df['Age'].mean())
```
