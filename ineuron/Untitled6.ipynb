```python
import re
import pandas as pd
```

```python
f= open('whatsappChatWChurail.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python

#pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s(?:am|pm)\s-\s'
pattern = r'\d{1,2}\/\d{1,2}\/\d{2,4},\s\d{1,2}:\d{2}\s[ap]m'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates= re.findall(pattern, data)
dates
```

```python
#df= pd.DataFrame({'user_messages':messages, 'message_date': dates})
#convert message_data type
#df['message_data'] = pd.to_datetime(df['message_date'], format= '%d/%m/%Y, %I:%M %p' )

#df.rename(columns={'message_date': 'date'}, inplace=True)

df= pd.DataFrame({'user_messages':messages, 'message_date': dates})
#convert message_data type
df['message_data'] = pd.to_datetime(df['message_date'], format= '%d/%m/%Y, %I:%M %p')

df.rename(columns={'message_date': 'date'}, inplace=True)
```
